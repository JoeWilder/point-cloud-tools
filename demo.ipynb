## Augmentations

```python
from src.utilities.augmentations import random_scale_point_cloud, random_sample, random_point_dropout, jitter_point_cloud

plot_point_cloud(pcd, title="Original")

sampled_pcd = random_sample(pcd)
plot_point_cloud(sampled_pcd, title="Random Sampling")

scaled_pcd = random_scale_point_cloud(pcd)
plot_point_cloud(scaled_pcd, title="Random Scaling")

dropped_pcd = random_point_dropout(pcd)
plot_point_cloud(dropped_pcd, title="Random Dropout")

jittered_pcd = jitter_point_cloud(pcd)
plot_point_cloud(jittered_pcd, title="Jittered")

```
